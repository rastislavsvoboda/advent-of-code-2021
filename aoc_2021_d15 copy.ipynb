```python
from collections import defaultdict, deque, Counter
import heapq

# pypy3.exe .\save.py 15

lines = open('15.in').readlines()
# lines = open('15.ex1').readlines()

M = []
for line in lines:
    line = line.strip()
    M.append([int(c) for c in line])
# print(M)


def solve1(maze):
    res = None
    M = maze
    R = len(M)
    C = len(M[0])
    start_risk = 0
    start_pos = (0, 0)
    end_pos = (R-1, C-1)
    heap = []
    heapq.heappush(heap, (start_risk, start_pos))
    DR = [-1, 0, 1, 0]
    DC = [0, 1, 0, -1]
    seen = set()
    while len(heap):
        (risk, pos) = heapq.heappop(heap)
        seen.add(pos)

        if pos == end_pos:
            res = risk
            break

        r, c = pos
        for i in range(4):
            rr = r + DR[i]
            cc = c + DC[i]
            if 0 <= rr < R and 0 <= cc < C:
                newPos = (rr, cc)
                if newPos not in seen:
                    heapq.heappush(heap, (risk+M[rr][cc], (rr, cc)))
    return res


def risk2(r, c, M, R, C):
    # each time the tile repeats to the right or downward,
    # all of its risk levels are 1 higher than the tile immediately up or left of it.
    # However, risk levels above 9 wrap back around to 1.
    newRisk = M[r % R][c % C] + r // R + c // C
    if newRisk > 9:
        newRisk -= 9
    return newRisk


def solve2(maze):
    res = None
    M = maze
    R = len(M)
    C = len(M[0])
    # the entire cave is actually five times larger in both dimensions
    RT = R * 5
    CT = C * 5
    start_risk = 0
    start_pos = (0, 0)
    end_pos = (RT-1, CT-1)
    heap = []
    heapq.heappush(heap, (start_risk, start_pos))
    DR = [-1, 0, 1, 0]
    DC = [0, 1, 0, -1]
    seen = set()
    while len(heap):
        risk, pos = heapq.heappop(heap)
        if pos in seen:
            continue
        seen.add(pos)


        if pos == end_pos:
            res = risk
            break

        r, c = pos
        for i in range(4):
            rr = r + DR[i]
            cc = c + DC[i]
            if 0 <= rr < RT and 0 <= cc < CT:
                newPos = (rr, cc)
                # if newPos not in seen:
                heapq.heappush(heap, (risk2(rr, cc, M, R, C), newPos))
    return res


print(solve1(M))
print(solve2(M))
```
